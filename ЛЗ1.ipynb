2. Сумма a и b, возведённая в квадрат:

```python
a = int(input("Введите число a: "))
b = int(input("Введите число b: "))
сумма_в_квадрате = (a + b) ** 2
print("Сумма, возведённая в квадрат:", сумма_в_квадрате)
```

4. Формула квадрата суммы чисел:

```python
a = int(input("Введите число a: "))
b = int(input("Введите число b: "))
квадрат_суммы = (a + b) ** 2
print(f"Формула квадрата суммы: ({a} + {b})^2 = {квадрат_суммы}")
```

5. Площадь и периметр неравнобедренного треугольника:

```python
a = float(input("Введите длину стороны a: "))
b = float(input("Введите длину стороны b: "))
c = float(input("Введите длину стороны c: "))
периметр = a + b + c
print("Периметр треугольника:", периметр)
p = периметр / 2
площадь = (p * (p - a) * (p - b) * (p - c)) ** 0.5
print("Площадь треугольника:", площадь)
```

6. Решение неравенства:

```python
a = int(input("Введите число a: "))
b = int(input("Введите число b: "))
результат = ((a**3 + 14) / 5) * b
print("Результат неравенства:", результат)
```

7. Квадрат числа, деление без остатка:

```python
a = int(input("Введите число a: "))
n = int(input("Введите число n: "))
результат = (a ** 2) // n
print(f"Квадрат числа {a}, делённый на {n} без остатка:", результат)
```

8. Операции с числами:

```python
a = float(input("Введите первое дробное число: "))
b = float(input("Введите второе дробное число: "))
print("Результат деления без остатка:", a // b)
a = int(input("Введите первое натуральное число: "))
b = int(input("Введите второе натуральное число: "))
print("Результат умножения:", a * b)
a = int(input("Введите первое натуральное число: "))
b = int(input("Введите второе натуральное число: "))
print("Остаток от деления:", a % b)
```

Задачи повышенной сложности:

1. Перевод секунд в дни, часы и минуты:

```python
n = int(input("Введите количество секунд: "))
дни = n // 86400
часы = (n % 86400) // 3600
минуты = (n % 3600) // 60
секунды = n % 60
print(f"{дни} дней, {часы} часов, {минуты} минут и {секунды} секунд")
```

2. Число n + nn + nnn:

```python
n = input("Введите число n: ")
результат = int(n) + int(n*2) + int(n*3)
print("Результат выражения n + nn + nnn:", результат)
```
