1. Календарь: программа выводит название месяца по его номеру:

```python
months = ["Январь", "Февраль", "Март", "Апрель", "Май", "Июнь", "Июль", "Август", "Сентябрь", "Октябрь", "Ноябрь", "Декабрь"]
number = int(input("Введите номер месяца (от 1 до 12): "))
if 1 <= number <= 12:
    print(f"Месяц: {months[number - 1]}")
else:
    print("Неверный номер месяца.")
```

4. Проверка високосного года:

```python
def is_year_leap(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False
year = int(input("Введите год: "))
print(is_year_leap(year))
```

5. Проверка простого числа

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input("Введите число: "))
print(is_prime(n))
```

6. Обмен значениями переменных

```python
def swap_if_conditions_met(a, b):
    условие1 = a > 3.6 * b
    условие2 = (-138 / 2) ** 1.3 <= b <= abs((-69 / (28**5.1)) * 4)
    if условие1 or условие2:
        a, b = b, a
    print("Значение a:", a, "Значение b:", b)
a = float(input("Введите значение a: "))
b = float(input("Введите значение b: "))
swap_if_conditions_met(a, b)
```

7. Проверка уникальности и подсчёт характеристик чисел

```python
def analyze_numbers():
    numbers = [int(input(f"Введите число {i + 1}: ")) for i in range(5)]
    unique = len(set(numbers)) == 5
    even_count = sum(1 for num in numbers if num % 2 == 0)
    negative_count = sum(1 for num in numbers if num < 0)
    in_range_count = sum(1 for num in numbers if -256 <= num <= 1024)
    print(f"Все числа уникальны: {unique}")
    print(f"Количество чётных чисел: {even_count}")
    print(f"Количество отрицательных чисел: {negative_count}")
    print(f"Количество чисел в диапазоне [-256, 1024]: {in_range_count}")
analyze_numbers()
```

Задание повышенной сложности
Решение неравенства и проверка чётности результата

```python
def solve_inequality_and_check_even(a, b, c):
    try:
        result = ((a ** 2 + b ** 3) / (c + 3)) / 4
        print(f"Результат неравенства: {result}")
        if result % 2 == 0:
            print("Результат чётный.")
        else:
            print("Результат нечётный.")
    except ZeroDivisionError:
        print("Ошибка: деление на ноль!")
a = float(input("Введите значение a: "))
b = float(input("Введите значение b: "))
c = float(input("Введите значение c: "))
solve_inequality_and_check_even(a, b, c)
```
